```python
import numpy as np
import pylab as plt

# Interval variabel t
# Membuat 500 titik data dari 0 sampai 2π
# Titik-titik ini digunakan sebagai domain (sumbu-x) untuk fungsi
t = np.linspace(0, 2*np.pi, 500)

# Definisi fungsi x(t)
# Rumus: x(t) = 2*cos(2t) - sin(t/2)
x = 2*np.cos(2*t) - np.sin(t/2)

# Membuat figure dengan ukuran 8x5 inch
plt.figure(figsize=(8,5))

# Membuat plot garis untuk x(t) terhadap t
# Label menggunakan LaTeX agar rumus tampil lebih rapi di legenda
plt.plot(t, x, label=r'$x(t) = 2\cos(2t) - \sin(\frac{t}{2})$')

# Memberi label sumbu-x
plt.xlabel('t')

# Memberi label sumbu-y
plt.ylabel('x(t)')

# Judul grafik
plt.title('Grafik fungsi x(t)')

# Menambahkan grid agar grafik lebih mudah dibaca
plt.grid(True)

# Menampilkan legenda (label fungsi)
plt.legend()

# Menampilkan grafik ke layar
plt.show()
```

```python
import numpy as np
import pylab as plt

# Interval variabel t
# Membuat 500 titik data antara 0 sampai 10
# Digunakan sebagai domain (sumbu-x) untuk fungsi
t = np.linspace(0, 10, 500)

# Definisi fungsi x(t)
# Rumus: x(t) = 4 * e^(-t/2)
# Fungsi ini adalah eksponensial menurun (decaying exponential)
x = 4 * np.exp(-t/2)

# Membuat figure dengan ukuran 8x5 inch
plt.figure(figsize=(8,5))

# Membuat plot garis untuk x(t) terhadap t
# Label menggunakan LaTeX agar rumus terlihat rapi di legenda
plt.plot(t, x, label=r'$x(t) = 4e^{-t/2}$')

# Memberi label pada sumbu-x
plt.xlabel('t')

# Memberi label pada sumbu-y
plt.ylabel('x(t)')

# Memberi judul grafik
plt.title('Sinyal x(t) = 4e^{-t/2}, 0 < t < 10')

# Menambahkan grid agar grafik lebih mudah dibaca
plt.grid(True)

# Menampilkan legenda (label fungsi)
plt.legend()

# Menampilkan grafik ke layar
plt.show()
```

```python
import numpy as np
import pylab as plt

# Interval variabel t
# Membuat 500 titik dari 0 sampai 2π
t = np.linspace(0, 2*np.pi, 500)

# Fungsi Heaviside (step function)
# H(t - π) = 0 untuk t < π, dan H(t - π) = 1 untuk t ≥ π
H = np.heaviside(t - np.pi, 1)

# Definisi sinyal x(t)
# x(t) = 3 * sin(3t) * H(t - π)
# Artinya: sinyal sinusoidal 3*sin(3t) hanya aktif mulai t ≥ π
x = 3 * np.sin(3*t) * H

# Membuat figure dengan ukuran 8x5 inch
plt.figure(figsize=(8,5))

# Membuat plot sinyal x(t) terhadap t
# Label ditulis dalam format LaTeX agar terlihat rapi
plt.plot(t, x, label=r'$x(t) = 3 \sin(3t)H(t-\pi)$')

# Memberi label sumbu-x
plt.xlabel('t')

# Memberi label sumbu-y
plt.ylabel('x(t)')

# Judul grafik
plt.title('Sinyal x(t) dengan Step Function')

# Menambahkan grid agar grafik mudah dibaca
plt.grid(True)

# Menampilkan legenda (keterangan fungsi)
plt.legend()

# Menampilkan grafik
plt.show()
```

```python
import numpy as np
import pylab as plt

# Interval variabel t
# Membuat 500 titik dari 0 sampai 2
t = np.linspace(0, 2, 500)

# Definisi sinyal kompleks
# Rumus: x(t) = 3 * e^(j(2t + π/4))
# Dua cara penulisan setara:
# x = 3 * np.exp(1j*(2*t) + 1j*(np.pi/4))
# x = 3 * np.exp(1j*(2*t + np.pi/4))   ← versi ringkas
x = 3 * np.exp(1j*(2*t + np.pi/4))

# Ekstrak bagian real, imajiner, dan magnitudo dari x(t)
x_real = np.real(x)   # bagian real
x_imag = np.imag(x)   # bagian imajiner
x_abs  = np.abs(x)    # magnitudo (panjang vektor kompleks)

# Membuat figure dengan ukuran 8x5 inch
plt.figure(figsize=(8,5))

# Plot bagian real (garis penuh)
plt.plot(t, x_real, label='real(x(t))')

# Plot bagian imajiner (garis putus-putus)
plt.plot(t, x_imag, '--', label='imag(x(t))')

# Plot magnitudo (garis titik-titik)
plt.plot(t, x_abs, ':', label='|x(t)|')

# Label sumbu
plt.xlabel('t')
plt.ylabel('x(t)')

# Judul plot dengan LaTeX agar rumus rapi
plt.title(r'$x(t) = 3 e^{j(2t + \pi/4)}, \; 0<t<2$')

# Tambahkan legenda dan grid
plt.legend()
plt.grid(True)

# Tampilkan plot
plt.show()

# (Opsional) menampilkan beberapa nilai awal
# untuk melihat isi array t, real(x), imag(x), dan |x|
print('t[0:5] =', t[:5])
print('real(x)[0:5] =', np.round(x_real[:5], 4))
print('imag(x)[0:5] =', np.round(x_imag[:5], 4))
print('|x|[0:5] =', np.round(x_abs[:5], 4))
```

```python
import numpy as np
import pylab as plt

# Parameter sampling
dt = 0.001                     # resolusi waktu (step sampling)
t = np.arange(-2, 3 + dt, dt)  # vektor waktu dari -2 s/d 3 dengan langkah dt

# Inisialisasi sinyal nol (array dengan panjang sama seperti t)
x = np.zeros_like(t)

# Cari indeks terdekat untuk t=0 dan t=1
# Tujuannya supaya bisa menempatkan impuls tepat di titik waktu tersebut
idx0 = np.argmin(np.abs(t - 0.0))
idx1 = np.argmin(np.abs(t - 1.0))

# Set nilai impuls
# Nilai 1/dt dipakai agar luas area impuls (nilai * dt) ≈ 1
x[idx0] = 1.0 / dt   # impuls di t=0
x[idx1] = 1.0 / dt   # impuls di t=1

# Plot dengan stem agar impuls terlihat jelas sebagai garis vertikal
plt.figure(figsize=(8,4))
plt.stem(t, x, linefmt='C0-', markerfmt='C0o', basefmt='k-')
plt.xlim(-2, 3)                     # batasi sumbu-x
plt.ylim(0, 1.2*(1.0/dt))           # batasi sumbu-y dengan margin 20% di atas nilai impuls
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title(r'Approximation: $x(t)=\delta(t)+\delta(t-1)$  pada  $-2 \leq t \leq 3$')
plt.grid(True)
plt.show()

# Verifikasi area impuls (harus mendekati 1)
# Menghitung integral aproksimasi: nilai impuls * dt ≈ 1
area0 = x[idx0] * dt
area1 = x[idx1] * dt
print(f'Area impuls di t=0 ≈ {area0:.6f}')
print(f'Area impuls di t=1 ≈ {area1:.6f}')

```
